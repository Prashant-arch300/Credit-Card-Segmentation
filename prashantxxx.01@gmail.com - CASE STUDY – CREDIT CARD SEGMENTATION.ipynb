### 5.Payments to minimum payments ratio 

```python
credit['pay_min_ratio']=credit.PAYMENTS/credit.MINIMUM_PAYMENTS
```

```python
credit.describe()
```

```python
def purchase(credit):
    
   
    if (credit['ONEOFF_PURCHASES']==0) & (credit['INSTALLMENTS_PURCHASES']==0):
        return 'none'
    if (credit['ONEOFF_PURCHASES']>0) & (credit['INSTALLMENTS_PURCHASES']>0):
         return 'both_oneoff_installment'
    if (credit['ONEOFF_PURCHASES']>0) & (credit['INSTALLMENTS_PURCHASES']==0):
        return 'one_off'
    if (credit['ONEOFF_PURCHASES']==0) & (credit['INSTALLMENTS_PURCHASES']>0):
        return 'istallment'
    
```

```python
credit['purchase_type']=credit.apply(purchase,axis=1)
```

```python
def outlier_capping(x):
    x = x.clip_upper(x.quantile(0.99))
    x = x.clip_lower(x.quantile(0.01))
    return x
```

```python
credit_new=credit.drop(['CUST_ID','purchase_type'],axis=1)
```

```python
# outlier treatment
credit_new=credit_new.apply(outlier_capping)
```

## Clustering

# f). Clustering 

---
<big>

Based on the intuition on type of purchases made by customers and their distinctive behavior exhibited based on the purchase_type (as visualized above in Insights from KPI) , starting with 4 clusters.

</big>

---

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=4,random_state=123)
```

```python
km.fit(reduced_cr)
```

```python
km.labels_
```

```python
pd.Series(km.labels_).value_counts()
```

```python
cluster_range=range(1,21)
cluster_errors=[]
for num in cluster_range:
    cluster=KMeans(num)
    cluster.fit(reduced_cr)
    cluster_errors.append(cluster.inertia_)
```

```python
cluster_df=pd.DataFrame({'num':cluster_range,'cluster_errors':cluster_errors})
```

```python
cluster_df[0:21]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.grid(True)
plt.plot(cluster_df.num,cluster_df.cluster_errors,marker='o')
plt.figure(figsize=(12,6))
```

```python
from sklearn import metrics
```

```python
score=[]
k_range=range(3,10)
for k in k_range:
    km_score=KMeans(n_clusters=k,random_state=123)
    km_score.fit(reduced_cr)
    
    score.append(metrics.silhouette_score(reduced_cr,km_score.labels_))
```

```python
score
```

```python
plt.plot(k_range,score)
plt.grid(True)
```

```python
# Key performace variable selection . here i am dropping varibales which are used in derving new KPI
col_kpi=['PURCHASES_TRX','Monthly_Avg_Purchase','Monthly_cash_advance','Limit_usage','CASH_ADVANCE_TRX',
         'pay_min_ratio','istallment','one_off','none','CREDIT_LIMIT']
```

```python
credit
```

```python
cre_original=pd.concat([credit,pd.get_dummies(credit['purchase_type'])],axis=1)
```

```python
# Conactenating labels found through KMeans with data 
cluster_df_4=pd.concat([cre_original[col_kpi],pd.Series(km.labels_,name='Cluster_4')],axis=1)
```

```python
cluster_df_4.Cluster_4.value_counts()
```

```python
# Mean value gives a good indication of the distribution of data. So we are finding mean value for each variable for each cluster
cluster_4=cluster_df_4.groupby('Cluster_4')\
.apply(lambda x: x[col_kpi].mean()).T
```

```python
fig,ax=plt.subplots(figsize=(15,10))
index=np.arange(len(cluster_4.columns))

cash_advance=np.log(cluster_4.loc['Monthly_cash_advance',:].values)
credit_score=(cluster_4.loc['Limit_usage',:].values)
purchase= np.log(cluster_4.loc['Monthly_Avg_Purchase',:].values)
payment=cluster_4.loc['pay_min_ratio',:].values
istallment=cluster_4.loc['istallment',:].values
one_off=cluster_4.loc['one_off',:].values


bar_width=.10
b1=plt.bar(index,cash_advance,color='b',label='Monthly cash advance',width=bar_width)
b2=plt.bar(index+bar_width,credit_score,color='m',label='Credit_score',width=bar_width)
b3=plt.bar(index+2*bar_width,purchase,color='k',label='Avg purchase',width=bar_width)
b4=plt.bar(index+3*bar_width,payment,color='c',label='Payment-minpayment ratio',width=bar_width)
b5=plt.bar(index+4*bar_width,istallment,color='r',label='installment',width=bar_width)
b6=plt.bar(index+5*bar_width,one_off,color='g',label='One_off purchase',width=bar_width)

plt.xlabel("Cluster")
plt.title("Insights")
plt.xticks(index + bar_width, ('Cl-0', 'Cl-1', 'Cl-2', 'Cl-3'))
plt.legend()
```
